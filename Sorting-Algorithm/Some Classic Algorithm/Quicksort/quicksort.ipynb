```python
import numpy as np

class mysort:
    def list_generator(self,length=10,listtype=False):
        if listtype:
            self.display_list=np.linspace(1,length,length,dtype=int).tolist()
            # np.random.shuffle(self.display_list)
        else:
            self.display_list=np.linspace(1,length,length,dtype=int)
        np.random.shuffle(self.display_list)

        return self.display_list

    def _quicksort(self,nums,low,high):
        if high<=low:
            return
        pvoid = self.partition(nums,low,high)
        self._quicksort(nums,low,pvoid-1)
        self._quicksort(nums,pvoid+1,high)

    def partition(self,nums,low,high):
        pvoid = nums[high]
        i = low-1
        for j in range(low,high):
            if nums[j]<=pvoid:
                i+=1
                nums[i],nums[j] = nums[j],nums[i]
        i+=1
        nums[i],nums[high] = nums[high],nums[i]
        return i

    def quicksort(self,nums):
        self._quicksort(nums,0,len(nums)-1)
```

```python
test=mysort()
test.list_generator(180)
print('测试样例为：')
print(test.display_list)
test.quicksort(test.display_list)
print('排序结果为：')
print(test.display_list)
```
