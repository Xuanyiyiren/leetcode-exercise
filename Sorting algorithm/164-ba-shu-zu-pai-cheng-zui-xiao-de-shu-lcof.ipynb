# Problem: [LCR 164. 破解闯关密码](https://leetcode.cn/problems/ba-shu-zu-pai-cheng-zui-xiao-de-shu-lcof/description/)


# 思路1：

希望排列成最小的数，如果只有个位数，这等价于排序，将最小的排在前面。

如果是考虑多位数，需要寻找合适的判序规则，判断谁应该在前面。

但是判序规则似乎很复杂，与其思考判序规则应该如何，我们不妨直接实验。我们不妨更大胆一些，只考虑只有这两个数，那么需要组合顺序是它们的组合起来最小的顺序。在更一般的情形下，这个判序规则仍然应该成立！（除非这是一个偏序关系而非全序关系，但是偏序关系是无法构成链的，所以大可放心，这一定是全序关系）所以，判序规则一句话写完。

于是先写一个判序函数，然后利用这个函数排序即可，我这里抄了一个[快排](https://blog.csdn.net/weixin_43250623/article/details/88931925)。

```python
class Solution:
    # def key(self,num1:int,num2:int)->bool:
    #     str1=str(num1)
    #     str2=str(num2)
    #     len1=len(str1)
    #     len2=len(str2)
    #     flag=0
    #     for i in range(min(len1,len2)):
    #         if str1[i]<str2[i]:
    #             flag=1
    #             break
    #         elif str1[i]>str2[i]:
    #             flag=2
    #             break
    #     if flag==1:
    #         return True
    #     elif flag==2:
    #         return False
    #     else:
    #         if len1==len2:
    #             return True
    #         elif len1<len2:
    #             if str2[i+1]=='0':
    #                 return False
    #             else:
    #                 return True
    #         else:
    #             if str1[i+1]=='0':
    #                 return True
    #             else:
    #                 return False
    def key(self,num1:int,num2:int)->bool:
        return int(str(num1)+str(num2))<int(str(num2)+str(num1))
    
    def quick_sort(self,alist, start, end):
        """快速排序"""
        if start >= end:  # 递归的退出条件
            return
        mid = alist[start]  # 设定起始的基准元素
        low = start  # low为序列左边在开始位置的由左向右移动的游标
        high = end  # high为序列右边末尾位置的由右向左移动的游标
        while low < high:
            # 如果low与high未重合，high(右边)指向的元素大于等于基准元素，则high向左移动
            while low < high and not (self.key(alist[high], mid)):
                high -= 1
            alist[low] = alist[high]  # 走到此位置时high指向一个比基准元素小的元素,将high指向的元素放到low的位置上,此时high指向的位置空着,接下来移动low找到符合条件的元素放在此处
            # 如果low与high未重合，low指向的元素比基准元素小，则low向右移动
            while low < high and self.key(alist[low],mid):
                low += 1
            alist[high] = alist[low]  # 此时low指向一个比基准元素大的元素,将low指向的元素放到high空着的位置上,此时low指向的位置空着,之后进行下一次循环,将high找到符合条件的元素填到此处

        # 退出循环后，low与high重合，此时所指位置为基准元素的正确位置,左边的元素都比基准元素小,右边的元素都比基准元素大
        alist[low] = mid  # 将基准元素放到该位置,
        # 对基准元素左边的子序列进行快速排序
        self.quick_sort(alist, start, low - 1)  # start :0  low -1 原基准元素靠左边一位
        # 对基准元素右边的子序列进行快速排序
        self.quick_sort(alist, low + 1, end)  # low+1 : 原基准元素靠右一位  end: 最后

    def numlist_to_str(self,nums:list[int]) -> str:
        s=''
        for i in nums:
            s+=str(i)
        return s

    def crackPassword(self, password: list[int]) -> str:
        sorted_password=password.copy()
        self.quick_sort(sorted_password,0,len(password)-1)
        return self.numlist_to_str(sorted_password)
```

```python
test = Solution()
test1 = [15,8,7]
print('第一个测试用例为:')
print(test1)
# print(test.movelastzero(test1))
ans1=test.crackPassword(test1)
print('测试结果为:')
print(ans1)

print()
# test = Solution()
test2 = [0, 3, 30, 34, 5, 9]
print('第二个测试用例为:')
print(test2)
# print(test.movelastzero(test1))
ans2=test.crackPassword(test2)
print('测试结果为:')
print(ans2)

print()
# test = Solution()
test3 = [121,12]
print('第三个测试用例为:')
print(test3)
# print(test.movelastzero(test1))
ans3=test.crackPassword(test3)
print('测试结果为:')
print(ans3)
```
